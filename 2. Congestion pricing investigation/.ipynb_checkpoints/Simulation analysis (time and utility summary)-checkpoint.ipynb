# By vehicle rate

```python
def grouped_bar_plot_by_rate(df, ax, measure, vehicle_rate):
    """
    Plots the given data in a grouped bar chart that groups data by each scenario.
    
    Parameters:
        df (pandas DataFrame): a dataframe that has summarized data categorized into
                               different scenarios (each row represents a scenario) and
                               different cases (each column represents a case)
    """
    
    # Group by each scenario
    df.plot(x='Scenario',
            kind='bar',
            stacked=False,
            ax=ax,
            color=["#66c2a5", '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854'],
            figsize=(12,6),
            rot=30,
            width=0.6)
    
    # Change scenario names on the x axis
    ax.set_xticklabels([r"Lower $\eta$", r"Higher $\eta$", "Average scenario", 
                    r"Lower $\gamma$", r"Higher $\gamma$"])
    
    # Add other features
    ax.set_ylabel(f"Average {measure} per vehicle")
    ax.set_title(f"Average {measure} per vehicle at vehicle rate = {vehicle_rate}")
    plt.legend([r"No connection, $\alpha=0$", r"With connection, $\alpha=0$", 
                r"With connection, Lower $\alpha$", r"With connection, Average $\alpha$", 
                r"With connection, Higher $\alpha$"], bbox_to_anchor=(1, 0.75))
    
    return ax
```

## Vehicle rate = 22

```python
fig, ax = plt.subplots()
ax = grouped_bar_plot_by_rate(time_22, ax, "travel time", "22")

plt.savefig("Analysis/Rate22_Time.png",
            bbox_inches = 'tight', pad_inches=0.25)
plt.show()
```

```python
fig, ax = plt.subplots()
ax = grouped_bar_plot_by_rate(utility_22, ax, "utility", "22")
ax.set_yscale("symlog")

plt.savefig("Analysis/Rate22_Utility.png",
            bbox_inches = 'tight', pad_inches=0.25)
plt.show()
```

# Improvment plots
In this part, only the cases with positive penalty factor $\alpha$ will be plotted. Their values will be expressed as a ratio when compared to the case of no congestion pricing with middle connection. For example, if for a certain positive penalty factor $\alpha$ the average travel time is $T_1$, and the corresponding average time for the case of no congestion pricing with middle connection is $T_2$, then the value of improvement is computed as $\frac{T_2}{T_1}-1$, with a positive value representing that imposing a posivie $\alpha$ helps reduce the travel time, and vice versa.

```python
time_22 = pd.read_excel("2.1 Vehicle rate = 22/Summary_22.xlsx", sheet_name="Overall_time")
utility_22 = pd.read_excel("2.1 Vehicle rate = 22/Summary_22.xlsx", sheet_name="Overall_utility")

time_27 = pd.read_excel("2.2 Vehicle rate = 27/Summary_27.xlsx", sheet_name="Overall_time")
utility_27 = pd.read_excel("2.2 Vehicle rate = 27/Summary_27.xlsx", sheet_name="Overall_utility")

time_45 = pd.read_excel("2.3 Vehicle rate = 45/Summary_45.xlsx", sheet_name="Overall_time")
utility_45 = pd.read_excel("2.3 Vehicle rate = 45/Summary_45.xlsx", sheet_name="Overall_utility")
```

```python
def improvement_bar_by_rate(df, ax, measure, vehicle_rate):
    imp_df = pd.DataFrame()
    imp_df["Scenario"] = df["Scenario"]
    
    # Compute the improvement ratio
    for c in df.columns[3:6].tolist():
        imp_df[c+"_improvement"] = df.Zero_alpha / df[c] - 1
    
    imp_df.plot(x='Scenario',
             kind='bar',
             stacked=False,
             ax=ax,
             color=["#66c2a5", '#fc8d62', '#8da0cb'],
             figsize=(10,6),
             rot=0,
             width=0.5)
    
    ax.set_xticklabels([r"Lower $\eta$", r"Higher $\eta$", "Average scenario", 
                    r"Lower $\gamma$", r"Higher $\gamma$"])
    
    ax.set_ylabel(f"Improvement in {measure} by ratio")
    ax.set_title(f"Improvement in average {measure} per vehicle at vehicle rate = {vehicle_rate}")
    plt.legend([r"With connection, Lower $\alpha$",
                r"With connection, Average $\alpha$", 
                r"With connection, Higher $\alpha$"], bbox_to_anchor=(1, 0.75))
    return ax
```

## Vehicle rate = 22

```python
# Improvement in time
fig, ax = plt.subplots()
improvement_bar_by_rate(time_22, ax, "time", "22")

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.1%}'.format(x) for x in vals])

plt.savefig("Analysis/Rate22_Time_Improve.png",
            bbox_inches = 'tight', pad_inches=0.25)
plt.show()

# Improvement in utility
fig, ax = plt.subplots()
improvement_bar_by_rate(utility_22, ax, "utility", "22")

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.1%}'.format(x) for x in vals])

plt.savefig("Analysis/Rate22_Utility_Improve.png",
            bbox_inches = 'tight', pad_inches=0.25)
plt.show()
```
